# Distributions

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
```

### Bernoulli Distribution

```python
ks = [0, 1]
p = 0.5
bern = stats.bernoulli(p)
```

```python
plt.stem(ks, bern.pmf(ks))
```

```python
plt.stem(ks, bern.cdf(ks))
```

### Binomial distribution

```python
n = 10
ks = range(0, n)
p = 0.5
bin = stats.binom(p=p, n=n)
```

```python
plt.stem(ks, bin.pmf(ks))
```

```python
plt.stem(ks, bin.cdf(ks))
```

### Poisson distribution

```python
lmbda = 2
poisson = stats.poisson(mu=lmbda)
ks = range(0, int(5 * lmbda + 1))
```

```python
plt.stem(ks, poisson.pmf(ks))
```

```python
plt.stem(ks, poisson.cdf(ks))
```

### Gaussian distribution

```python
mu = 5
stddev = 3
xs = np.linspace(mu - 5 * stddev, mu + 5 * stddev, 200)
norm = stats.norm(loc=mu, scale=stddev)
```

```python
plt.plot(xs, norm.pdf(xs))
```

```python
plt.plot(xs, norm.cdf(xs))
```
